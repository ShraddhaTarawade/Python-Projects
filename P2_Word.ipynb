```python
search_word_count = input('Enter the word: ')
search_word = search_word_count.lower()
file = open("Word.txt", "r")
read_data = file.read()
word_count = read_data.lower().count(search_word)
print(f"The word '{search_word_count}' appeared {word_count} times.")
```

```python
from collections import Counter 
# Accepting input from the user 
user_response_file_name = input("Please provide the filename with extension : ") 
try :
# Reading the file content and creating a words LIST 
    def read_words(user_response_file_name): 
        with open(user_response_file_name, 'r') as file_object: 
            words = [] 
            for line in file_object: 
                words += line.split() 
                return words 
        # Calling definition created above 
    words_list = read_words(user_response_file_name) 
        # Using package collections to count words 
    count_each_word = Counter(words_list) 
    print(count_each_word) 
except OSError : 
    print("The filename provided by you is incorrect");
```
